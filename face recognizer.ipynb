Finding encodings of the face to map images

```python
faceLoc = face_recognition.face_locations(imgtrain)[0]
encodetrain = face_recognition.face_encodings(imgtrain)[0]
cv2.rectangle(imgtrain,(faceLoc[3],faceLoc[0]),(faceLoc[1],faceLoc[2]),(255,0,255),2) # top, right, bottom, left
 
faceLocTest = face_recognition.face_locations(imgTest)[0]
encodeTest = face_recognition.face_encodings(imgTest)[0]
cv2.rectangle(imgTest,(faceLocTest[3],faceLocTest[0]),(faceLocTest[1],faceLocTest[2]),(255,0,255),2)
```
